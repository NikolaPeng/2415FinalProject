```python
import pandas as pd
# Load the dataset
data = pd.read_csv("Dataset.csv")
# Display the first few rows of the dataset
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(3, 1, figsize=(14, 18))
# Scatter plot for Bitcoin vs NYSE
sns.regplot(x=data["BTC"], y=data["NYSE"], ax=ax[0], color="green", scatter_kws={'s':10}, line_kws={"color": "red"})
ax[0].set_title("Bitcoin vs NYSE Closing Prices")
ax[0].set_xlabel("Bitcoin Closing Price")
ax[0].set_ylabel("NYSE Closing Price")
# Scatter plot for Bitcoin vs NASDAQ
sns.regplot(x=data["BTC"], y=data["NASDAQ"], ax=ax[1], color="blue", scatter_kws={'s':10}, line_kws={"color": "red"})
ax[1].set_title("Bitcoin vs NASDAQ Closing Prices")
ax[1].set_xlabel("Bitcoin Closing Price")
ax[1].set_ylabel("NASDAQ Closing Price")
# Scatter plot for Bitcoin vs LSE
sns.regplot(x=data["BTC"], y=data["LSE"], ax=ax[2], color="purple", scatter_kws={'s':10}, line_kws={"color": "red"})
ax[2].set_title("Bitcoin vs LSE Closing Prices")
ax[2].set_xlabel("Bitcoin Closing Price")
ax[2].set_ylabel("LSE Closing Price")
# display
plt.tight_layout()
plt.show()
fig.savefig("scatter_plots.png", bbox_inches='tight')
```

```python
# Calculating the correlation matrix for the prices and volumes
correlation_matrix = data[["BTC", "NYSE", "NASDAQ", "LSE", "BTC_Volume", "NYSE_Volume", "NASDAQ_Volume", "LSE_Volume"]].corr()
# Setting the figure size
plt.figure(figsize=(12, 8))
# Generating the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1, linewidths=.5)
# Formatting the visualization
plt.title("Correlation Heatmap of Market Prices and Volumes")
plt.tight_layout()
# Display the heatmap
plt.savefig("heatmap.png", bbox_inches='tight')
plt.show()
```
